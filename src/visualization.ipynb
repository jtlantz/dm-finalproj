```python
import pandas as pd
import numpy as np
import re
import nltk
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer

"""
If you use the VADER sentiment analysis tools, please cite:

Hutto, C.J. & Gilbert, E.E. (2014). VADER: A Parsimonious Rule-based Model for
Sentiment Analysis of Social Media Text. Eighth International Conference on
Weblogs and Social Media (ICWSM-14). Ann Arbor, MI, June 2014.
"""
nltk.download('vader_lexicon')
```

```python
# Load back in sentiment score df
df1 = pd.read_csv('../resources/sentiment_scores.csv')
df2 = pd.read_csv('../resources/raw-dataset.csv')

df = pd.merge(df1, df2, on='row_id')
```

```python
# Plot the sentimen_score vs the rating on a violin plot
fig, ax = plt.subplots(figsize=(10, 6))
ax.violinplot([df[df.rating == 1].compound, df[df.rating == 2].compound, df[df.rating == 3].compound, df[df.rating == 4].compound, df[df.rating == 5].compound], showmeans=True, showmedians=True)
ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xticklabels(['1', '2', '3', '4', '5'])
ax.set_xlabel('Rating')
ax.set_ylabel('Sentiment Score')
ax.set_title('Sentiment Score vs Rating')
plt.show()
```

```python
# Same violin plot but only for ones with a label OR

fig, ax = plt.subplots(figsize=(10, 6))
ax.violinplot([df[(df.rating == 1) & (df.label == 'OR')].compound, df[(df.rating == 2) & (df.label == 'OR')].compound, df[(df.rating == 3) & (df.label == 'OR')].compound, df[(df.rating == 4) & (df.label == 'OR')].compound, df[(df.rating == 5) & (df.label == 'OR')].compound], showmeans=True, showmedians=True)
ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xticklabels(['1', '2', '3', '4', '5'])
ax.set_xlabel('Rating')
ax.set_ylabel('Sentiment Score')
ax.set_title('Sentiment Score vs Rating, REAL Review Only')
plt.show()
```

```python
# Same violin plot but only for ones with a label OR

fig, ax = plt.subplots(figsize=(10, 6))
ax.violinplot([df[(df.rating == 1) & (df.label == 'CG')].compound, df[(df.rating == 2) & (df.label == 'CG')].compound, df[(df.rating == 3) & (df.label == 'CG')].compound, df[(df.rating == 4) & (df.label == 'CG')].compound, df[(df.rating == 5) & (df.label == 'CG')].compound], showmeans=True, showmedians=True)
ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xticklabels(['1', '2', '3', '4', '5'])
ax.set_xlabel('Rating')
ax.set_ylabel('Sentiment Score')
ax.set_title('Sentiment Score vs Rating, FAKE Review Only')
plt.show()
```
