```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
df = pd.read_csv('../resources/sentiment_scores.csv')
df
```

```python
df['target'] = np.where(df['label']=='CG', 1, 0)
```

```python
X = df['text']
y = df['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, shuffle=True)
```

```python
bundled_pipeline = Pipeline([("tfidf", TfidfVectorizer()), ("svc", LinearSVC())])
bundled_pipeline.fit(X_train, y_train)
y_pred = bundled_pipeline.predict(X_test)
my_accuracy_score = accuracy_score(y_test, y_pred)
my_precision_score = precision_score(y_test, y_pred)
my_recall_score = recall_score(y_test, y_pred)
my_roc_auc_score = roc_auc_score(y_test, y_pred)
print('Accuracy:', my_accuracy_score)
print('Precision:', my_precision_score)
print('Recall:', my_recall_score)
print('ROC/AUC:', my_roc_auc_score)
```

```python
bundled_pipeline = Pipeline([("tfidf", TfidfVectorizer()), ("random_forest", RandomForestClassifier())])
bundled_pipeline.fit(X_train, y_train)
y_pred = bundled_pipeline.predict(X_test)
my_accuracy_score = accuracy_score(y_test, y_pred)
my_precision_score = precision_score(y_test, y_pred)
my_recall_score = recall_score(y_test, y_pred)
my_roc_auc_score = roc_auc_score(y_test, y_pred)
print('Accuracy:', my_accuracy_score)
print('Precision:', my_precision_score)
print('Recall:', my_recall_score)
print('ROC/AUC:', my_roc_auc_score)
```

```python
bundled_pipeline = Pipeline([("tfidf", TfidfVectorizer()), ("decision_tree", DecisionTreeClassifier())])
bundled_pipeline.fit(X_train, y_train)
y_pred = bundled_pipeline.predict(X_test)
my_accuracy_score = accuracy_score(y_test, y_pred)
my_precision_score = precision_score(y_test, y_pred)
my_recall_score = recall_score(y_test, y_pred)
my_roc_auc_score = roc_auc_score(y_test, y_pred)
print('Accuracy:', my_accuracy_score)
print('Precision:', my_precision_score)
print('Recall:', my_recall_score)
print('ROC/AUC:', my_roc_auc_score)
```

```python
bundled_pipeline = Pipeline([("tfidf", TfidfVectorizer()), ("k_neighbors", KNeighborsClassifier())])
bundled_pipeline.fit(X_train, y_train)
y_pred = bundled_pipeline.predict(X_test)
my_accuracy_score = accuracy_score(y_test, y_pred)
my_precision_score = precision_score(y_test, y_pred)
my_recall_score = recall_score(y_test, y_pred)
my_roc_auc_score = roc_auc_score(y_test, y_pred)
print('Accuracy:', my_accuracy_score)
print('Precision:', my_precision_score)
print('Recall:', my_recall_score)
print('ROC/AUC:', my_roc_auc_score)
```
